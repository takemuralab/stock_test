### 複数社のデータをまとめる

```python
appl = yf.Ticker('AAPL')
days = 20 
hist = appl.history(period=f'{days}d')
hist.index = hist.index.strftime('%d %B %Y')
hist = hist[['Close']]
hist.columns = ['apple']
hist = hist.T
hist.index.name = 'Name'
hist
```

```python
days = 20

tickers = {
    'apple': 'AAPL',
    'meta': 'META',
    'google' :  'GOOGL',
    'microsoft' : 'MSFT',
    'netflix' : 'NFLX',
    'amazon' : 'AMZN'
}

#company = 'meta'

def get_data(days, tickers):
    df = pd.DataFrame()

    for company in tickers.keys():
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
        
    return df
```

```python
days = 20
df = get_data(days, tickers)
df
```

### グラフの可視化　（Altair はstreamlitに親和性が高いライブラリ）

#### 下準備

```python
!pip3 install altair
```

```python
import altair as alt
```

```python
df.head()
```

```python
companies = ['apple', 'meta']
data = df.loc[companies]
data
```

```python
data.sort_index()
```

```python
#altairはpandasのデータを引っ張ってこれる
data = data.T.reset_index()
data.head()
```

```python
#日付　データ　株価の列に変更したい
data = pd.melt(data, id_vars=['Date']).rename(
    columns={'value': 'Stock Prices(USD)'}
)
data.head()
```

### グラフの可視化

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None),
        color='Name:N'
    )
)
```

```python
chart
```

### グラフのスケールの調整

```python
#scaleを chat内部に表記することで範囲を指定できる.　ただし、範囲外のものがあるときに注意が必要
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None, scale=alt.Scale(domain=[200,250])),
        color='Name:N'
    )
)
chart
```

```python
#範囲外のグラフを排除, mark_lineの内部に clip=Trueを追加
ymin, ymax = 200, 250

　
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None, scale=alt.Scale(domain=[ymin, ymax])),
        color='Name:N'
    )
)
chart
```
